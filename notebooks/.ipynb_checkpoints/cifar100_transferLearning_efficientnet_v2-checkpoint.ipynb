```python
import tensorflow as tf
from tensorflow.keras import layers, Sequential
from tensorflow.keras.callbacks import ReduceLROnPlateau, EarlyStopping
import matplotlib.pyplot as plt
from tensorflow.keras.applications.efficientnet_v2 import preprocess_input
from sklearn.model_selection import train_test_split
```

```python
tf.__version__
```

```python
#load Dataset
(X_train_full, y_train_full), (X_test, y_test) = tf.keras.datasets.cifar100.load_data()
```

```python
X_train, X_val, y_train, y_val = train_test_split(
    X_train_full, y_train_full, test_size=0.2, random_state=42, stratify=y_train_full
)
```

```python
import numpy as np

print("y_train shape:", y_train.shape)
print("num classes:", len(np.unique(y_train)))
print("min/max label:", y_train.min(), y_train.max())
```

```python
# Convert NumPy arrays into a tf.data pipeline (yields (image, label) pairs)

ds_train = tf.data.Dataset.from_tensor_slices((X_train, y_train))
ds_val = tf.data.Dataset.from_tensor_slices((X_val, y_val))
ds_test = tf.data.Dataset.from_tensor_slices((X_test, y_test))
```

```python
# Data augmentation
def data_augment(image, label):
    image = tf.cast(image, tf.float32) / 255.0
    image = tf.image.random_brightness(image, max_delta=0.2)
    image = tf.image.random_hue(image, max_delta=0.2)
    image = tf.image.random_contrast(image, lower=0.8, upper=1.2)
    image = tf.image.random_flip_left_right(image)
    image = tf.image.random_saturation(image, lower=0.8, upper=1.2)

    image = tf.clip_by_value(image, 0.0, 1.0)
    image = image * 255.0

    return image, label
```

```python
IMG_SIZE = 240
def preprocessing(image, label):
    image = tf.cast(image, tf.float32)
    image = tf.image.resize(image, (IMG_SIZE, IMG_SIZE))
    image = preprocess_input(image)
    label = tf.squeeze(label, axis=-1)
    return image, label
```

```python
# prepare pipline
def prepare_pip(ds, augment=False, shuffle=False):
    ds = ds.shuffle(10000) if shuffle else ds
    ds = ds.map(data_augment, num_parallel_calls=tf.data.AUTOTUNE) if augment else ds
    ds = ds.map(preprocessing, num_parallel_calls=tf.data.AUTOTUNE)
    ds = ds.batch(20)
    ds = ds.prefetch(tf.data.AUTOTUNE)
    return ds
```

```python
# Preprocess the data
# Preprocess the data
train_ds = prepare_pip(ds_train, augment=True, shuffle=True)
val_ds = prepare_pip(ds_val, augment=False, shuffle=False)
test_ds = prepare_pip(ds_test, augment=False, shuffle=False)
```

## Implement *"efficientnet_v2"* using transfer Learning

```python

input_shape = (240, 240, 3)

base_model = tf.keras.applications.EfficientNetV2B0(include_top=False, weights='imagenet',input_shape=input_shape,pooling='avg')

base_model.trainable = False

model = tf.keras.Sequential(
    [
    tf.keras.layers.Input(shape=input_shape),
    base_model,
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(128, activation="relu", kernel_initializer="he_normal"),
    tf.keras.layers.Dropout(rate=0.2),
    tf.keras.layers.Dense(100, activation="softmax")
])

model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=0.001),
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"],
)

model.summary()
```

```python
reduce_lr = ReduceLROnPlateau(monitor="val_accuracy", factor=0.2, patience=5, min_lr=1e-6)
early_stopping = EarlyStopping(
    monitor="val_accuracy", patience=8, restore_best_weights=True
)
callbacks = [early_stopping, reduce_lr]
```

```python
# Train the model
history = model.fit(train_ds, epochs=100,callbacks=callbacks, validation_data=val_ds)
```

```python
# Plot training & validation accuracy values
plt.figure(figsize=(12, 4))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])
plt.title('Model accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

# Plot training & validation loss values
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])
plt.title('Model loss')
plt.ylabel('Loss')
plt.xlabel('Epoch')
plt.legend(['Train', 'Validation'], loc='upper left')

plt.tight_layout()
plt.show()
```

```python
# ۱. باز کردن قفل مدل پایه (Unfreeze)
# با این کار تمام لایه‌های EfficientNet آماده آموزش و تغییر می‌شوند
base_model.trainable = True

# ۲. نگاه داشتن قفل لایه‌های ابتدایی (اختیاری اما به شدت توصیه شده)
# لایه‌های اولیه معمولاً ویژگی‌های عمومی (مثل خطوط و لبه‌ها) را تشخیص می‌دهند که نیازی به تغییر ندارند.
# ما فقط می‌خواهیم لایه‌های انتهایی مدل پایه (که ویژگی‌های پیچیده‌تر را می‌فهمند) تغییر کنند.
# در اینجا مثلاً ۱۰۰ لایه اول را همچنان فریز نگه می‌داریم:
for layer in base_model.layers[:100]:
    layer.trainable = False

# ۳. کامپایل مجدد مدل (بسیار مهم!)
# هر زمان که وضعیت trainable لایه‌ها را تغییر می‌دهید، باید مدل را دوباره compile کنید.
# نکته حیاتی: نرخ یادگیری (learning_rate) باید بسیار کوچک باشد (مثلا 1e-5).
# اگر نرخ یادگیری بزرگ باشد، وزن‌های عالیِ از پیش‌آموزش‌دیده خراب می‌شوند!
model.compile(
    optimizer=tf.keras.optimizers.Adam(learning_rate=1e-5),  # نرخ یادگیری بسیار کوچک
    loss=tf.keras.losses.SparseCategoricalCrossentropy(),
    metrics=["accuracy"],
)

# چاپ خلاصه مدل برای اطمینان از اینکه تعداد Trainable params افزایش یافته است
print("مدل پس از باز شدن قفل لایه‌ها:")
model.summary()

# ۴. ادامه آموزش مدل
# فرض می‌کنیم می‌خواهیم ۲۰ دور (Epoch) دیگر مدل را به صورت ظریف آموزش دهیم.
fine_tune_epochs = 20

# اگر مدل قبلی متوقف شده باشد، می‌فهمیم در چه دوری متوقف شده است
# تا شماره دورها در نمودار به درستی ادامه پیدا کند
if 'history' in locals():
    # پیدا کردن آخرین دوری که آموزش دیده
    last_epoch = history.epoch[-1] + 1
else:
    last_epoch = 50 # اگر متغیر history در دسترس نبود، فرض می‌کنیم 50 دور بوده

total_epochs = last_epoch + fine_tune_epochs

print(f"شروع Fine-tuning از دور {last_epoch} تا {total_epochs}...")

history_fine = model.fit(
    train_ds,
    epochs=total_epochs,
    initial_epoch=last_epoch,  # شروع آموزش از همان جایی که قبلاً قطع شده بود
    validation_data=val_ds,
    callbacks=callbacks  # همان توابع توقف زودهنگام و کاهش نرخ یادگیری قبلی
)
```
